```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips =sns.load_dataset("tips")
print(tips)
```

```python
sns.set_style('whitegrid')
sns.lineplot(data=tips,
             x='total_bill',
              y='tip'   )
```

```python
sns.set_style('whitegrid')
tipgraph = sns.lineplot(data=tips,
             x='total_bill',
              y='tip'   )
tipgraph.set(title = 'Total Bill',
             xlabel ='Total $',
             ylabel ='Tip Amount')
```

```python
fig = plt.subplots(figsize=(10,3))

sns.set_style('whitegrid')

tipgraph = sns.lineplot(data=tips,
             x='total_bill',
             y='tip' ,
             color='pink',
             linestyle='--',
             linewidth=3,
             marker='o',
             markerfacecolor='indigo'
                 )
tipgraph.set(title = 'Total Bill',
             xlabel ='Total $',
             ylabel ='Tip Amount')
```

```python
fig = plt.subplots(figsize=(10,3))

sns.set_style('whitegrid')

tipgraph = sns.scatterplot(data=tips,
             x='total_bill',
             y='tip' ,
             style='time',
             hue ='day',
             palette=['pink','skyblue', 'yellowgreen', 'purple']
                              )
tipgraph.set(title = 'Total Bill',
             xlabel ='Total $',
             ylabel ='Tip Amount')
```

```python
sns.set_style('whitegrid')

tipgraph = sns.pairplot(
            data=tips,
             hue = 'day' ,
              )
```

```python
daygraph = sns.relplot(
            data=tips,
            x='total_bill',
            y='tip',
            hue='sex',   # Color based on sex
            col='day',  # the day is the coloumn
            kind='scatter',
            col_wrap=2  # Two Columns only
              )
```

```python
import plotly.graph_objects as go
x1 = np.array(['Bash','Sam', 'Moh', 'Saleem'])
y1 = np.array([110, 180, 240, 99])
```

```python
graph=go.Figure()
graph.add_trace(go.Bar(x=x1, y=y1))
graph
```

```python
graph=go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1))
x=x1,
y=y1,
mode='markers',
marker=dict(
    size=15,
    color='hotpink',
    opacity=1,
    line=dict(width=5,
              color='purple')

)

graph.update(title='Interactive',
             xaxis_title='Pirate',
             yaxis_title='Scores',
             width=5,
             height=20)
```

```python
from wordcloud import WordCloud

df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv')

df
```

```python
text = " ".join(each for each in df.quote)

wordcloud = WordCloud(background_color= 'white',
                      colormap='inferno').generate(text)
fig, ax = plt.subplots(figsize=(7,3))
ax.imshow(wordcloud,
          interpolation='bilinear')
ax.axis("off")
```

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(['apple', 'Orange', 'banna', 'Tomato'])
B = set(['watermelon','strabery', 'apple', 'key'])
```

```python
diagram = venn2([A,B],
                set_labels = ('Fruit', 'Pies'),
                set_colors=('blue', 'red'),
                alpha =0.5)

diagram.get_label_by_id("10").set_text("\n".join(A-B))
diagram.get_label_by_id("11").set_text("\n".join(A&B))
diagram.get_label_by_id("01").set_text("\n".join(B-A))
```
