```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```
ax, fig = plt.subplots()
plt.plot(x,y1, label="Person 1")
plt.plot(x,y2,label="Person 2")
plt.legend(loc="upper left", bbox_to_anchor=(1,1),frameon = True, fontsize =12, ncol=1, shadow=True)
plt.show()
```

```
ax, fig = plt.subplots()
plt.scatter(x,y1, label="Person 1")
plt.scatter(x,y2,label="Person 2")
plt.legend(loc="lower left", frameon = True, fontsize =12, ncol=1, shadow=True)
plt.text(10,95, "This Value is Important", ha='center', color='red', size=15)
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
#fig, ax = plt.subplot()
ax.axis([0,10,0,10])
ax.text(1,5,". Data: (1,5)", transform=ax.transData)
ax.text(0.5,0.1,". Axes: (0.5,0.1)", transform=ax.transAxes)
ax.text(0.2,0.2,". Figure: (0.2,0.2)", transform=fig.transFigure)
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
plt.scatter(x,y1)
plt.scatter(x,y2)
ax.annotate("This is Important", xy=(10,95),
            xytext=(20,94),
            arrowprops=dict(facecolor='hotpink', arrowstyle="wedge"))

plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
plt.scatter(x,y1, label="Person 1")
plt.scatter(x,y2, label="Person 2")

plt.legend(loc="lower right")

ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))
ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
plt.scatter(x,y1, label="Person 1")
plt.scatter(x,y2, label="Person 2")

plt.legend(loc="lower right")

ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
plt.scatter(x,y1, label="Person 1")
plt.scatter(x,y2, label="Person 2")

plt.legend(loc="lower right")

ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
plt.scatter(x,y1, label="Person 1")
plt.scatter(x,y2, label="Person 2")

plt.legend(loc="lower right")

ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))
ax.tick_params(axis='x', rotation=45, labelsize=10)
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
plt.scatter(x,y1, label="Person 1")
plt.scatter(x,y2, label="Person 2")

plt.legend(loc="lower right")

ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))
ax.tick_params(axis='x', rotation=45, labelsize=10)
plt.xlabel("X-Axis", fontsize=18, fontweight='bold')
plt.ylabel("Y-Axis", fontsize=18, fontweight='bold')
plt.show()
```
